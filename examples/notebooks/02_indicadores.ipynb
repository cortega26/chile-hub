# Indicadores economicos

Carga la serie historica y obtiene el ultimo valor disponible por indicador.

```python
from chile_hub import ChileHub

indicadores = ChileHub().load_polars("indicadores")
```

```python
ultimos = (
    indicadores.sort("fecha", descending=True)
    .group_by("codigo_indicador")
    .first()
    .select("codigo_indicador", "fecha", "valor")
    .sort("codigo_indicador")
)
ultimos
```
